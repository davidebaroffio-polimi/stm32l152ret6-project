```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#df_nopass = pd.read_csv("faults_nopass_names.csv")

df_eddi_cfcss = pd.read_csv("faults_eddi_cfcss_names.csv")
df_eddi_rasm = pd.read_csv("faults_eddi_rasm_names.csv")
df_eddi_rasm_inter = pd.read_csv("faults_eddi_inter-rasm_names.csv")

df_seddi_cfcss = pd.read_csv("faults_seddi_cfcss_names.csv")
df_seddi_rasm = pd.read_csv("faults_seddi_rasm_names.csv")
df_seddi_rasm_inter = pd.read_csv("faults_seddi_inter-rasm_names.csv")

df_fdsc_cfcss = pd.read_csv("faults_fdsc_cfcss_names.csv")
df_fdsc_rasm = pd.read_csv("faults_fdsc_rasm_names.csv")
df_fdsc_rasm_inter = pd.read_csv("faults_fdsc_inter-rasm_names.csv")


dataframes = {"eddi_cfcss": df_eddi_cfcss,
              "eddi_rasm": df_eddi_rasm,
              "eddi_rasm_inter": df_eddi_rasm_inter,
              "fdsc_cfcss": df_fdsc_cfcss,
              "fdsc_rasm": df_fdsc_rasm,
              "fdsc_rasm_inter": df_fdsc_rasm_inter,
              "seddi_cfcss": df_seddi_cfcss,
              "seddi_rasm": df_seddi_rasm,
              "seddi_rasm_inter": df_seddi_rasm_inter
              }
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key].fillna("<UNDEF>")
```

```python
for key in dataframes:
    print(dataframes[key][dataframes[key]["struct"] == "<UNDEF>"]["struct"])
```

```python
df_new = pd.DataFrame()

for key in dataframes:
    df = dataframes[key]
    sz_tot = df.shape[0]
    df_errcodes = df.groupby('code')
    sz_df_errcodes = df_errcodes.size()
    df_new[key] = sz_df_errcodes

df_new
```
