```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_nopass = pd.read_csv("faults_nopass_names.csv")

df_eddi_cfcss = pd.read_csv("faults_eddi_cfcss_names.csv")
df_eddi_rasm = pd.read_csv("faults_eddi_rasm_names.csv")
df_eddi_rasm_inter = pd.read_csv("faults_eddi_inter-rasm_names.csv")

df_seddi_cfcss = pd.read_csv("faults_seddi_cfcss_names.csv")
df_seddi_rasm = pd.read_csv("faults_seddi_rasm_names.csv")
df_seddi_rasm_inter = pd.read_csv("faults_seddi_inter-rasm_names.csv")

df_fdsc_cfcss = pd.read_csv("faults_fdsc1_cfcss_names.csv")
df_fdsc_rasm = pd.read_csv("faults_fdsc1_rasm_names.csv")
df_fdsc_rasm_inter = pd.read_csv("faults_fdsc_inter-rasm_names.csv")


dataframes = {"nopass": df_nopass,
              "eddi_cfcss": df_eddi_cfcss,
              "eddi_rasm": df_eddi_rasm,
              "eddi_rasm_inter": df_eddi_rasm_inter,
              "fdsc_cfcss": df_fdsc_cfcss,
              "fdsc_rasm": df_fdsc_rasm,
              "fdsc_rasm_inter": df_fdsc_rasm_inter,
              "seddi_cfcss": df_seddi_cfcss,
              "seddi_rasm": df_seddi_rasm,
              "seddi_rasm_inter": df_seddi_rasm_inter
              }
```

```python
for key in dataframes:
    dataframes[key] = dataframes[key].fillna("<UNDEF>")
```

```python
df_new = pd.DataFrame()

df_new["nopass"] = pd.Series(data={-3: 0, -2: 0, -1: 0, 0: 0, 1: 0, 2:0})

for key in dataframes:
    df = dataframes[key]
    sz_tot = df.shape[0]
    df_errcodes = df.groupby('code')
    sz_df_errcodes = df_errcodes.size()
    df_new[key] = sz_df_errcodes

df_new
```

# Most vulnerable data structures

```python
df_names = pd.DataFrame()

for key in dataframes:
    df = dataframes[key]
    df_errcodes = df.groupby('struct')
    sz_df_errcodes = df_errcodes.size()
    df = pd.DataFrame()
    df[key] = sz_df_errcodes
    df_names = df_names.join(df, how="outer")

df_names.to_csv("faults_structs.csv")
```

```python
df_errs = pd.DataFrame()

for key in dataframes:
    df = dataframes[key]
    df = df[df["code"] == -2]
    df_errcodes = df.groupby('struct')
    sz_df_errcodes = df_errcodes.size()
    df = pd.DataFrame()
    df[key] = sz_df_errcodes
    df_errs = df_errs.join(df, how="outer")

df_errs = df_errs.fillna(0)
```

```python
(df_errs).dropna(how="all").replace(0, "NaN").round(4).to_csv("sdc_structs.csv")
```

# Link to location 

```python
loc_dict = dict()

loc_dict["gkvars"] = ["uxCurrentNumberOfTasks", "xTickCount", "uxTopReadyPriority", "xSchedulerRunning", "uxPendedTicks", "xYieldPending", "xNumOfOverflows", "uxTaskNumber", "xNextTaskUnblockTime", "uxSchedulerSuspended", "uxCurrentNumberOfTasks_dup", "xTickCount_dup", "uxTopReadyPriority_dup", "xSchedulerRunning_dup", "uxPendedTicks_dup",  "xYieldPending_dup", "xNumOfOverflows_dup", "uxTaskNumber_dup", "xNextTaskUnblockTime_dup", "uxSchedulerSuspended_dup"]

loc_dict["tcbvars"] = ["TCB"]

loc_dict["dldlst"] = ["pxDelayedTaskList", "pxDelayedTaskList_dup", "pxOverflowDelayedTaskList", "pxOverflowDelayedTaskList_dup", "xDelayedTaskList1", "xDelayedTaskList1_dup", "xDelayedTaskList2", "xDelayedTaskList2_dup"]

loc_dict["rdylst"] = ["pxReadyTasksLists", "pxReadyTasksLists_dup"]

loc_dict["mtxqvars"] = ["xMessageBuffer", "xMessageBuffer_dup", "globalQueuehandle", "ucStaticTimerQueueStorage", "xQueueRegistry", "xStaticQueue", "xStaticTimerQueue", "xTimerQueue"]

loc_dict["tmrvars"] = ["pxTimerBuffer", "pxTimerBuffer_dup", "pxCurrentTimerList", "pxCurrentTimerList_dup", "pxOverflowTimerList", "pxOverflowTimerList_dup", "xActiveTimerList1", "xActiveTimerList2", "xActiveTimerList1_dup", "xActiveTimerList2_dup"]

loc_dict["zOthers"] = [""]
```

```python
locations = ["None"] * df_names.shape[0]
i = 0

for row in df_names.index:
    found = False
    for key in loc_dict:
        for elem in loc_dict[key]:
            if elem in row:
                locations[i] = key
                found = True
        if found:
            break
    i = i+1
```

```python
df_names["location"] = locations
```

```python
df_names[df_names["location"] != "None"].groupby("location").sum()
```

```python
locations = ["None"] * df_errs.shape[0]
i = 0

for row in df_errs.index:
    found = False
    for key in loc_dict:
        for elem in loc_dict[key]:
            if elem in row:
                locations[i] = key
                found = True
        if found:
            break
    i = i+1

df_errs["location"] = locations
(df_errs[df_errs["location"] != "None"].groupby("location").sum() / df_names[df_names["location"] != "None"].groupby("location").sum() * 100).to_csv("loop_faults.csv")
```
